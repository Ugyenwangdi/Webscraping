## 1. Sending get requests

```python
url = "https://www.imdb.com/search/title/?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012"
response = requests.get(url)
print(response.url)
```

```python
moviesSoup = BeautifulSoup(response.text)

data_dictionary = {}

for movie in moviesSoup.findAll('div','lister-item mode-advanced'):
  title = movie.findAll('a')[1].contents[0]
  genres = movie.find('span','genre').contents
  runtime = movie.find('span',attrs={'class':'runtime'}).contents[0]
  rating = movie.find('span','value').contents[0]
  description = movie.findAll('p','text-muted')[1].contents
  voteGross = movie.findAll('span',attrs={'name':'nv'})
  votes = voteGross[0].contents[0]
  gross = voteGross[1].contents[0]

  data_dictionary[title] = [genres,runtime,rating,description,votes,gross]
  print(title,genres,runtime,rating,description,votes,gross)
print(data_dictionary)
```
