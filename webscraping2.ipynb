Given URL is the wikipedia page for Billboard Year-End Hot 100 singles of 1970. The page contain HTML table with 100 singles as shown below. 
image.png

**Task:**

*   Perform Web scraping and extract all 100 singles with their artists.

*   Create python dictionary which contains key as title of the single and value as lists of artists.


Note: Prepare detail documentation of the procedure with comments for code used.

```python
# Import Libraries
import requests
from bs4 import BeautifulSoup
```

```python
# sending request
url = "https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_1970"
response = requests.get(url)
print(response.url) # print url
response # response status
```

```python
# TASK:
# 1. Extract all 100 songs with their artists
# 2. Create Python dictionary which contains key as title and value as artists

songSoup = BeautifulSoup(response.text) # Object of BeautifulSoup

data_dictionary = {}

for song in songSoup.findAll('tr')[1:101]: # loop over index 1 to 101 because the findAll('tr') contains table headers
  # Priting 100 table rows.............
  # print(song)   

  title = song.findAll('a')[0].string

  artist = song.findAll('a')[1].string
  # Printing Titles and Artists.............
  print(title, ',', artist)

  # Printing Dictionary.............
  data_dictionary[title] = [artist]
print(data_dictionary)
```
