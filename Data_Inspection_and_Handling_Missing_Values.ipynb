**5. Handling Missing Value (Numerical**

**Droping missing values**

```python
#dropna() drops any row containing missing values. 
newdf = data.dropna()
```

```python
newdf.shape
```

```python
filtered_df = data[data.notnull()]
filtered_df.shape
```

```python
newdf1 = data.dropna(axis = 1)
newdf1.head()
```

**Filling Missing value**

```python
# using fillna() without any paramters
data.fillna(0)
```

```python
#Calling fillna with a dict you can use a different fill value for each column
data.fillna({"take_out":'False','review_count':10,'stars':3})[:50]
```

```python
# dataframe.bfill () is used to backward fill the missing values in the dataset. It will backward fill the NaN values that are present in the pandas dataframe
data.fillna(method='bfill')
data.bfill()[:100]
```

```python
# The ffill () method replaces the NULL values with the value from the previous row (or previous column, if the axis parameter is set to 'columns' )
data.fillna(method = 'ffill')
data.ffill()[:100]
```

```python
data.fillna({'take_out':'Empty','review_count':data['review_count'].mean()})
```

```python
# interpolate values
data.interpolate()
```

**6 Handling missing values(Categorical)**


**Deleting values**


```python
# Drop an observation or row. Drops second and third row
data.drop([1,2])
```

```python
# Drop a row by condition, drops name with China Sea Chinese Restaurant
data[data.name != 'China Sea Chinese Restaurant']
```

**Replacing with mode**


```python
data['review_count'] = data['review_count'].fillna(data['review_count'].mode()[0])
```
